```python
# In a new Code block, write a function that takes as its parameter a list of strings. The function should return
# True if one of the strings is coconut (in lower case, as shown), or False otherwise. Include some code
# outside the function that shows that the function works correctly. Again, you may assume the "Happy Path",
# so the list will always contain just strings.

def coconut(words):
    return "coconut" in words


# Code to show the function works
fruits1 = ["apple", "banana", "coconut", "orange"]
fruits2 = ["apple", "banana", "orange"]

print(coconut(fruits1))  # Expected: True
print(coconut(fruits2))  # Expected: False
```
